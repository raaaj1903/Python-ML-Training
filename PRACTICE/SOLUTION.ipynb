```python
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset( 'iris')
```

```python
df.sample(10)
```

```python
# 1--> Relationship Between Petal Length and Petal Width Across Species
# - Plot the relationship between petal_length and petal_width for different species.
# - Use axis-level (sns.scatterplot) and figure-level (sns.relplot) versions.
# - Highlight species differences using hue='species'

# --> axis level
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df,x="petal_length",y="petal_width",hue="species")
plt.title("Relationship Between Petal Length and Petal Width Across Species")
plt.show
```

```python

# 2->> figure level
sns.relplot(
    data=df,
    x='petal_length',
    y='petal_width',
    hue='species',
    kind='scatter',
    height=5,
    aspect=1.2
)
plt.suptitle("Petal Length vs Petal Width (Figure-Level)", y=1.02)
plt.show()

```

```python
#Trend of Sepal Length by Species
# - Show how average sepal_length varies across species.
# - Use line plots with sns.lineplot and sns.relplot.
# - Add error bars (standard deviation) to represent variability

#  1. Axis-Level Plot using sns.lineplot

plt.figure(figsize=(8,5))
sns.lineplot(data=df, x="species",y="sepal_length",errorbar="sd")
plt.title("Average Sepal Length by Species (with Std Dev)")
plt.ylabel("Sepal Length (cm)")
plt.xlabel("Species")
plt.grid(True)


plt.show
```

```python
#  2. Figure-Level Plot using sns.relplot
sns.relplot(
    data=df,
    x='species',
    y='sepal_length',
    kind='line',
    errorbar='sd',
    height=5,
    aspect=1.2
)
plt.suptitle("Average Sepal Length by Species (Figure-Level)", y=1.02)
plt.show()
```
