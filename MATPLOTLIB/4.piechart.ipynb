```python
#pie chart
import matplotlib.pyplot as plt
```

```python
x=[10,20,39,50]
l=['c','c++','python','java']

plt.pie(x)
plt.show
```

```python
plt.pie(x,labels=l)
plt.show
```

```python

#expload function
ex=[0.2,0.0,0.0,0.0]
plt.pie(x,labels=l, explode=ex)
plt.show
```

```python
#change color of pie chart

c=['r','g','y','b']
plt.pie(x,labels=l ,colors=c,)
plt.show
```

```python
#addd shadow 
ex=[0.2,0.0,0.0,0.0]
c=['r','g','y','b']
plt.pie(x,labels=l ,colors=c ,shadow=True,explode=ex)
plt.show
```

```python
# radius size of pie chart
plt.pie(x,labels=l ,colors=c,radius=1 )
plt.show
```
