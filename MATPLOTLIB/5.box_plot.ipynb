```python
import matplotlib.pyplot as plt
```

```python
#box plot
x=[10,20,30,40,50,60,70]
```

```python
plt.boxplot(x)
plt.show
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,widths=2)
plt.show
```

```python
#horixontslly
x=[10,20,30,40,50,60,70]
plt.boxplot(x,vert=False)
plt.show
```

```python
x=[10,20,30,40,50,60,70]
plt.boxplot(x,labels=['python'])
plt.show
```

```python
#multiplebox plot
x=[10,20,30,40,50,60,70]
y=[20,10,65,78,55,44,34]
z=[x,y]
plt.boxplot(z,tick_labels=["c++","python"])
plt.show
```
