```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]#x-data
y=[2,4,6,8,9,6,5]#y-data
plt.plot(x,y)#ploting the graph using x data and y data
```

```python
plt.plot(x,y)
plt.title("Numbers COUNT Frequency")
plt.xlabel('Numbers')
plt.ylabel("Frequncy")
plt.show # show plot
```

```python
#givin color to the line parameter
plt.plot(x,y ,color="red")#color give the specific line
plt.title("Numbers COUNT Frequency")
plt.xlabel('Numbers')
plt.ylabel("Frequncy")
plt.show # show plot
```

```python
#line style change dot 
plt.plot(x,y ,color="black",linestyle='dashdot')
plt.title("Numbers COUNT Frequency")
plt.xlabel('Numbers')
plt.ylabel("Frequncy")
plt.show # show plot
```

```python
#line width
plt.plot(x,y ,color="black",linestyle='dashdot' ,linewidth=2)
plt.title("Numbers COUNT Frequency")
plt.xlabel('Numbers')
plt.ylabel("Frequncy")
plt.show # show plot
```

```python
#line markesr in plot
plt.plot(x,y ,color="black",linestyle='dashdot' ,marker=">")
plt.title("Numbers COUNT Frequency")
plt.xlabel('Numbers')
plt.ylabel("Frequncy")
plt.show # show plot
```

```python
# multiline graph
x=[1,2,3,4,5,6,7]#x-data
y=[2,4,6,8,9,6,5]#y-data
z=[3,4,5,6,7,2,8]#z-data
plt.plot(x,y ,color="black",linestyle='dashdot' ,marker=">")
plt.title("Numbers COUNT Frequency")
plt.xlabel('Numbers')
plt.ylabel("Frequncy")
plt.plot(y,z ,color="red",linestyle='dashdot' ,marker=">")
plt.plot(x,z,color="blue",linestyle='dashdot' ,marker=">")
plt.show
```

```python
#label-
# legend()
plt.plot(x,y ,label="abc")
plt.plot(x,z, label="xyz")
plt.title("Numbers COUNT Frequency")
plt.xlabel('Numbers')
plt.ylabel("Frequncy")
plt.legend()
plt.show
```

```python
#using grid backg
plt.plot(x,y ,label="abc")
plt.plot(x,z, label="xyz")
plt.title("Numbers COUNT Frequency")
plt.xlabel('Numbers')
plt.ylabel("Frequncy")
plt.legend()
plt.grid()
plt.show
```
