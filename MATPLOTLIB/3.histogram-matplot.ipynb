```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
number=np.random.random(6)
plt.hist(number)
plt.show
```

```python
# labels
number=np.random.random(6)
plt.title("Frequence count")
plt.ylabel("count")
plt.xlabel("Nunmbers")
plt.hist(number)
plt.show
```

```python
#color edge color

number=np.random.random(6)
plt.hist("Frequence count", color="blue" ,edgecolor="red")
plt.ylabel("count")
plt.xlabel("Nunmbers")
plt.hist(number)
plt.show
```

```python
#count bins
number=np.random.random(6)
plt.hist("Frequence count", color="blue" ,edgecolor="red" ,bins=4)
plt.ylabel("count")
plt.xlabel("Nunmbers")
plt.hist(number)
plt.show
```

```python
number=np.random.random(6)
plt.hist("Frequence count", color="blue" ,edgecolor="red" )
plt.ylabel("count")
plt.xlabel("Nunmbers")
plt.hist(number)
plt.show
```
