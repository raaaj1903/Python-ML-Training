```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,5,7,8,9]
y=[3,2,4,5,6,7,1,24,5,4]
plt.scatter(x,y)
plt.show

#covert scaterplot to plot --># plt.plot(x,y)
```

```python
plt.scatter(x,y)
plt.title('corelation b/n x and y')
plt.xlabel('X')
plt.ylabel("Y")
plt.show # show plot
```

```python

plt.scatter(x,y ,color="black", marker="*", )
plt.title('corelation b/n x and y')
plt.xlabel('X')
plt.ylabel("Y")
plt.grid()
plt.show # show plot
```

```python
#bar-vertical
plt.bar(x,y)
plt.show()
```

```python
#bar-vertical
plt.bar(x,y ,color="yellow", )
plt.title("programing language vs popularity")
plt.xlabel("language")
plt.xlabel("popularity")
plt.show()
```

```python
plt.bar(x,y ,color="yellow", align="center" )
plt.title("programing language vs popularity")
plt.xlabel("language")
plt.xlabel("popularity")
plt.show()
```

```python
#edge color line width line style parameter
plt.bar(x,y ,color="yellow",width=0.4,edgecolor='r',linewidth=3,linestyle='-' )
plt.title("programing language vs popularity")
plt.xlabel("language")
plt.xlabel("popularity")
plt.show()
```

```python
#color shade alpha parameter
plt.bar(x,y ,color="yellow",alpha=0.1 )#color shade dark and low bar
plt.title("programing language vs popularity")
plt.xlabel("language")
plt.xlabel("popularity")
plt.show()
```

```python
#bar-horizontal graph
plt.barh(x,y ,color="yellow",label='popularity' )#label uses legend as a parameter
plt.title("programing language vs popularity")
plt.xlabel("language")
plt.xlabel("popularity")
plt.legend()
plt.show()
```
