two types of function
1-figure level function
2-axes level function

```python
import seaborn as sns
```

```python
df=sns.load_dataset('penguins')
#load penguins data set
```

```python
df.sample(10)
```

```python
sns.relplot(data=df,x="flipper_length_mm",y="bill_length_mm", hue='species')#find specific species

# print(df['species'].unique())
```

```python
#style parameter
sns.relplot(data=df,x="flipper_length_mm",y="bill_length_mm", hue='species',style="island")
```

```python
# size paremeter ke acording diffrece show karna
sns.relplot(data=df,x="flipper_length_mm",y="bill_length_mm", hue='species',size="body_mass_g")
```

```python
#figure level function--->relplot
sns.relplot(kind="line",data=df,x="flipper_length_mm",y="bill_length_mm", hue='species',size="body_mass_g")
# if you create line plot you use kind parameter and passs line as value
```

```python
#line plot drow directly
sns.lineplot(data=df,x="flipper_length_mm",y="bill_length_mm")
```

```python
#same operation perform in line plot

# hue parametter

sns.lineplot(data=df,x="flipper_length_mm",y="bill_length_mm", hue='species')
```

```python
#style parameter --->in line plot
sns.lineplot(data=df,x="flipper_length_mm",y="bill_length_mm", hue='species',style="island")
```

```python
# size paremeter ke acording diffrece show karna
sns.lineplot(data=df,x="flipper_length_mm",y="bill_length_mm", hue='species',size="body_mass_g")
```

```python
# new 

sns.relplot(kind="scatter",data=df,x="flipper_length_mm",y="bill_length_mm", hue='sex',size="body_mass_g",style="species",col="island")
```

```python
sns.relplot(kind="scatter",data=df,x="flipper_length_mm",y="bill_length_mm", hue='sex',size="body_mass_g",style="species",row="island")
```

```python
sns.relplot(kind="scatter",data=df,x="flipper_length_mm",y="bill_length_mm", hue='sex',size="body_mass_g",style="species",col="island",row="")
```

```python
#cool wrap paremeter 
sns.relplot(kind="scatter",data=df,x="flipper_length_mm",y="bill_length_mm", hue='sex',size="body_mass_g",style="species",col="island",col_wrap=2)
```
